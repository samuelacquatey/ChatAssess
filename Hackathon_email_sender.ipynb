```python
import pandas as pd

def generate_email(name, email, phone, participation, alx_status, source, questions, question_response):
    greeting = f"Hi {name}," if pd.notna(name) else "Hi,"

    confirmation_text = f"Thanks for signing up as a **{participation}**! We're thrilled to have you join us for this exciting event."

    event_details = f"""
**Event Details**
- **Opening Date:** 26th April 2025
- **Time:** 10 AM
- **Venue:** ALX Hub, 1 Airport Square, Accra (4th Floor)
    """

    hackathon_timeline = """
**Hackathon Timeline**
- **April 6** – You will receive the dataset and additional resources.
- **April 21** – Deadline for submitting your visualization.
- **April 24** – Announcement of the Top 10 visualizations.
- **April 25** – Announcement of the Top 5 finalists.
- **April 26** – Final presentations at ALX Hub, Accra.
    """

    visualization_requirements = """
**Visualization Requirements**
To qualify, your visualization must:
- Be interactive or static but clearly communicate insights creatively.
- Address illegal mining’s environmental impact.
- Include a brief write-up of *300 words or less* or a *video of not more than 7 minutes* explaining your findings (*A video presentation is preferred*).
- Be submitted as a link to an interactive dashboard.
- Include solutions for mitigating environmental damage caused by illegal mining.
    """

    judging_criteria = """
**Judging Criteria**
Your visualization will be evaluated based on:
- **Clarity** – Is the message clear, impactful, and interpretable by all?
- **Creativity** – How engaging and innovative is the visualization?
- **Relevance** – Does it effectively highlight key insights creatively?
- **Technical Execution** – Accuracy, effective storytelling, and overall presentation.
    """

    question_section = f"""
**Your Query:**
*“{questions}”*
**Response:** {question_response}
    """ if pd.notna(questions) else ""

    closing_text = "We’re excited to see your creativity in action. See you at the hackathon!"

    return f"""
{greeting}

{confirmation_text}

{event_details}

{hackathon_timeline}

{visualization_requirements}

{judging_criteria}

{question_section}

{closing_text}


Best,
**Samuel Kingsley Acquatey**
*Hackathon Organizer, The Open Data Bank Ghana *
    """.strip()

# Load CSV file
df = pd.read_csv("tester - Sheet1.csv", encoding="utf-8-sig")

# Ensure column names are correctly formatted
df.columns = df.columns.str.strip()

# Generate email responses
df["Email Response"] = df.apply(
    lambda row: generate_email(
        row["Name"],
        row["Email"],
        row["Phone Number"],
        row["Are you participating in the hackathon or attending as an audience member?"],
        row["Are you an ALX alumni or current learner?"],
        row["How did you hear about this event?"],
        row["Comments, Suggestions, Questions?"],
        row["Question Response"],
    ), axis=1
)

# Save updated CSV
df.to_csv("participants_with_emails.csv", index=False, encoding="utf-8-sig")
print("Emails generated successfully!")
```

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

# Function to format text with proper HTML
def format_to_html(text):
    html_text = text.replace("**", "<b>").replace("*", "<i>")
    html_text = html_text.replace("\n", "<br>")  # Preserve line breaks
    return html_text

# Function to send an email
def send_email(to_email, name, email_body):
    sender_email = "samuelacquatey@gmail.com"
    sender_password = "orjr kdrd fkie npjz"  # Use App Password, NOT your Gmail password

    # Email Setup
    msg = MIMEMultipart()
    msg["From"] = sender_email
    msg["To"] = to_email
    msg["Subject"] = "Welcome to the Galamsey Data Visualization Hackathon!"

    # Convert to HTML format
    html_body = f"""
    <html>
    <body>
        {format_to_html(email_body)}
    </body>
    </html>
    """

    msg.attach(MIMEText(html_body, "html"))

    try:
        # Connect to Gmail SMTP server
        server = smtplib.SMTP("smtp.gmail.com", 587)
        server.starttls()  # Secure the connection
        server.login(sender_email, sender_password)
        server.sendmail(sender_email, to_email, msg.as_string())
        server.quit()
        print(f"✅ Email sent successfully to {name} ({to_email})")
    except Exception as e:
        print(f"❌ Failed to send email to {name} ({to_email}): {e}")

# Load CSV file from Colab's directory
df = pd.read_csv("participants_with_emails.csv", encoding="utf-8-sig")

# Send emails to all participants
for index, row in df.iterrows():
    send_email(row["Email"], row["Name"], row["Email Response"])
```
